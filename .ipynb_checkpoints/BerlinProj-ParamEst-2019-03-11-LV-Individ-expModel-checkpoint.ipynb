## Setting Up Environment

```julia
# Determine the net growth rate/differential growth rate of fibroblast
function individLV(du,u,p,t)
    du[1] = p[1].*u[1]
end
```

```julia
param = [0.01]
u0 = [0.1]; tspan = (0.0,72.0); 

PROB = ODEProblem(individLV,u0,tspan,param);
sol = solve(ODEProblem(individLV,u0,tspan,param),Vern8(),abstol=1e-8,reltol=1e-8); 
```

```julia
function calcAICscore(RSS,n,k)
    AIC = 2.0*(k+1)+n+log(RSS/n)+n*(1+log(2*pi))
end
```

## MEF - Wildtype Fibroblasts

```julia
source = "./MEFonly_20180802.csv"
#source = "/home/4467528/2018-08-BerlinProj/MEFonly_20180802.csv"
datatableMEF = CSV.read(source; delim=',');
dataMEF = Matrix(datatableMEF);
```

```julia
# function to calculate the loss: loss(solve(diffeq), data)
function my_loss_func_MEF(sol)
    tot_loss = 0.0
    if any((s.retcode != :Success for s in sol.u))
      tot_loss = Inf
    else
        # calculation for the loss here
        for i in 1:length(sol.u)
            solI = sol.u[i]
            x = dataMEF[i,2]
            y = solI(72)[1]
            tot_loss += sum((abs(x .- y).^2)./((x).^2))
        end
    end
    tot_loss
end
```

```julia
# Set up Monte Carlo Problem
Nmef = size(dataMEF,1);
initial_conditionsMEF = Array{Any}(undef,Nmef,1);
for i=1:Nmef
   initial_conditionsMEF[i] = [Float64(dataMEF[i,1])];
end
function prob_funcMEF(prob,i,repeat)
  ODEProblem(prob.f,initial_conditionsMEF[i],prob.tspan,prob.p);
end
monte_probMEF = MonteCarloProblem(PROB,prob_func=prob_funcMEF);
simMEF = solve(monte_probMEF,Rodas5(),num_monte=Nmef,parallel_type=:threads);
```

```julia
errMEForg = my_loss_func_MEF(simMEF)
```

```julia
#cost_functionMEF = build_loss_objective(monte_probMEF, Vern8(), my_loss_func_MEF, num_monte=Nmef, verbose=true)
function build_regularized_loss_func(λ)
    function regularized_loss_func(odeparams) 
        #println("executing cost_func: $odeparams")
        function prob_funcMEF(prob,i,repeat)
            ODEProblem(prob.f,initial_conditionsMEF[i],prob.tspan,odeparams);
        end
        monte_probMEF = MonteCarloProblem(PROB,prob_func=prob_funcMEF);
        simMEF = solve(monte_probMEF,
            Rosenbrock23(autodiff=false), # LinearExponential #Rodas5(),#Vern8(),#Rosenbrock23(autodiff=false),#radau(), #CVODE_BDF(),
            num_monte=Nmef,
            parallel_type=:threads,
            force_dtmin=true,
           # maxiters = 1000000
            );
        return my_loss_func_MEF(simMEF) + λ*sum(abs.(odeparams));
    end
    return regularized_loss_func;
end
```

```julia
resultMEF = @time Optim.optimize(build_regularized_loss_func(0),
    [-5.0], [5.0], ## upper and lower bounds 
    param, 
    Fminbox(LBFGS()),
    Optim.Options(f_tol=0.0, g_tol=1e-16, allow_f_increases=true, iterations=100000))
```
